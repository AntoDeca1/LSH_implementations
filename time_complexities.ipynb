```python
import matplotlib.pyplot as plt
import time
import numpy as np
from tqdm import tqdm
import big_o as big_o
```

```python
def measure_time(func, args):
    start_time = time.time()
    # Execute the operation
    _ = func(*args) - 0.5
    end_time = time.time()
    return end_time - start_time
```

## np.random.rand time complexities estimate


```python
# Different sizes for testing
sizes = [(10, 10, 10), (100, 100, 100), (1000, 1000, 1000)]
trials = 50
times = []

# Measure time for each size
for size in sizes:
    temp_times = []
    for trial in tqdm(range(trials)):
        time_taken = measure_time(np.random.rand, size)
        temp_times.append(time_taken)
    times.append(np.mean(temp_times))

# Plotting the results
plt.plot([np.prod(size) for size in sizes], times, marker='o')
plt.xlabel('Total number of elements')
plt.ylabel('Time taken (seconds)')
plt.title('Time Complexity of np.random.rand - .5')
plt.grid(True)
plt.show()
```

```python
def matrix_mask(matrix):
    output_matrix = (matrix > 0).astype(int)
    return output_matrix
```

```python
matrices = [np.random.rand(10, 10, 10), np.random.rand(100, 100, 100), np.random.rand(200, 200, 200),
            np.random.rand(200, 200, 200), np.random.rand(300, 300, 300), np.random.rand(400, 400, 400),
            np.random.rand(500, 500, 500)]
trials = 100
times = []
for matrix in matrices:
    temp_times = []
    for trial in tqdm(range(trials)):
        time_taken = measure_time(matrix_mask, matrix)
        temp_times.append(time_taken)
    times.append(np.mean(temp_times))
```

```python
# Plotting the results
plt.plot([np.sum(matrix.shape) for matrix in matrices], times, marker='o')
plt.xlabel('Total number of elements')
plt.ylabel('Time taken (seconds)')
plt.title('Time Complexity of >0 + astype(int)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import time
import matplotlib.pyplot as plt
from tqdm import tqdm


def measure_unique_time(matrix):
    start_time = time.time()
    unique_rows = np.unique(matrix, axis=0)
    end_time = time.time()
    return end_time - start_time


# Different sizes for testing
m_values = [10, 50, 100, 200, 300, 400]
nbits_values = [10, 20, 30, 40, 50, 60]
trials = 10
empirical_times = []
theoretical_times = []

# Measure time for each combination of m and nbits
for m in m_values:
    for nbits in nbits_values:
        temp_times = []
        for _ in tqdm(range(trials)):
            matrix = np.random.randint(0, 2, size=(m, nbits))
            time_taken = measure_unique_time(matrix)
            temp_times.append(time_taken)
        avg_time = np.mean(temp_times)
        empirical_times.append((m, nbits, avg_time))
        # Calculate the theoretical complexity
        theoretical_time = m * np.log(m) * nbits
        theoretical_times.append((m, nbits, theoretical_time))

# Normalize the times for comparison
empirical_m_nbits_product = [m * nbits for m, nbits, _ in empirical_times]
empirical_avg_times = [time for _, _, time in empirical_times]
theoretical_m_nbits_product = [m * nbits for m, nbits, _ in theoretical_times]
theoretical_values = [time for _, _, time in theoretical_times]

# Normalize the times to the same scale
max_empirical_time = max(empirical_avg_times)
max_theoretical_value = max(theoretical_values)

normalized_empirical_times = [time / max_empirical_time for time in empirical_avg_times]
normalized_theoretical_values = [time / max_theoretical_value for time in theoretical_values]

plt.plot(empirical_m_nbits_product, normalized_empirical_times, marker='o', label='Normalized Empirical Times')
plt.plot(theoretical_m_nbits_product, normalized_theoretical_values, marker='x', linestyle='--',
         label='Normalized Theoretical Complexity (m log m * nbits)')
plt.xlabel('Product of m and nbits')
plt.ylabel('Normalized Time taken')
plt.title('Normalized Empirical vs Theoretical Time Complexity of np.unique')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import time
import matplotlib.pyplot as plt
from tqdm import tqdm

def stringify_array(array: np.array):
    return ''.join(map(str, array.flatten()))

def measure_stringify_time(array):
    start_time = time.time()
    _ = stringify_array(array)
    end_time = time.time()
    return end_time - start_time

# Different sizes for testing
array_sizes = [100, 1000, 5000, 10000, 20000, 50000, 100000]
trials = 10
empirical_times = []

# Measure time for each array size
for size in array_sizes:
    temp_times = []
    for _ in tqdm(range(trials)):
        array = np.random.randint(0, 100, size=size)
        time_taken = measure_stringify_time(array)
        temp_times.append(time_taken)
    avg_time = np.mean(temp_times)
    empirical_times.append(avg_time)

# Normalize the times to the same scale
max_empirical_time = max(empirical_times)
max_theoretical_time = max(theoretical_times)

normalized_empirical_times = [time / max_empirical_time for time in empirical_times]
normalized_theoretical_times = [time / max_theoretical_time for time in theoretical_times]

# Plot the normalized results
plt.plot(array_sizes, normalized_empirical_times, marker='o', label='Normalized Empirical Times')
plt.plot(array_sizes, normalized_theoretical_times, marker='x', linestyle='--', label='Normalized Theoretical O(n)')
plt.xlabel('Number of elements')
plt.ylabel('Normalized Time taken')
plt.title('Normalized Empirical vs Theoretical Time Complexity of stringify_array')
plt.legend()
plt.grid(True)
plt.show()

```
